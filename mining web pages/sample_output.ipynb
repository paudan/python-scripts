Automatic summarization

```python
from parse_feeds import summarize
import json
import codecs
fname = 'feed.json'
blog_data = json.loads(open(fname).read())
for post in blog_data:
    post.update(summarize(post['content']))
    print post['title']
    print '=' * len(post['title'])
    print
    print 'Top N Summary'
    print '-------------\n',
    print ' '.join([codecs.encode(_post, 'ascii', 'ignore') for _post in post['top_n_summary']])
    print '-------------------'
    print 'Mean Scored Summary'
    print '-------------------\n'
    print ' '.join([codecs.encode(_post, 'ascii', 'ignore') for _post in post['mean_scored_summary']])
    print
```

Discovering interactions between entities

```python
from parse_feeds import extract_interactions, visualize_interactions
import json
import codecs
fname = 'feed.json'
blog_data = json.loads(open(fname).read())
for post in blog_data:
    post.update(extract_interactions(post['content']))
    print post['title']
    print '-' * len(post['title'])
    for interactions in post['entity_interactions']:
        print '; '.join([codecs.encode(i[0], 'ascii', 'ignore') for i in interactions])
    print
visualize_interactions(fname)
```
