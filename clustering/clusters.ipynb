### Document clustering###

```python
from clusters import *
```

Generate document clusters:

```python
blognames, words, data = readfile('blogdata.txt')
clust = hcluster(data)
```

Output these clusters:

```python
printclust(clust,labels=blognames)
```

Draw dendrogram representing these clusters:

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
img = drawdendrogram(clust, blognames)
fig2 = plt.figure(figsize = (20,60))
ax3 = fig2.add_subplot(111)
ax3.imshow(img, interpolation='none')
plt.show()
img.save('blogclust.jpg', 'JPEG')
```

Represent documents visually in 2D space using MDS technique:

```python
coords = scaledown(data)
im = draw2d(coords, blognames)
fig3 = plt.figure(figsize = (25,25))
ax4 = fig3.add_subplot(111)
ax4.imshow(im)
plt.show()
im.save('blogs2d.jpg', 'JPEG')
```

Rotate the matrix and run the same operations for clustering and drawingthe dendrogram. Since the matrix has been rotated,the words rather than the blogs are now the labels

```python
rdata = rotatematrix(data)
wordclust = hcluster(rdata)
im = drawdendrogram(wordclust,labels=words)
fig3 = plt.figure(figsize = (15,150))
ax4 = fig3.add_subplot(111)
ax4.imshow(im)
plt.show()
im.save('wordclust.jpg', 'JPEG')
```
